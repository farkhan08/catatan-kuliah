Class vs Object

a. Class 
    
    Class adalah suatu tempat untuk membuat suatu objek. Sedangkan objek sendiri mempunyai variavel anggota yang dan perilaku yang sama dengan Class.
    Struktur data yang bisa kita gunakan untuk mendefinisikan objek yang menyimpan data bersama-sama nilai-nilai dan perilaku (behavior). Kelas adalah suatu entitas yang merupakan bentuk program dari suatu abstraksi untuk permasalahan dunia nyata, dan instans dari class merupakan realisasi dari beberapa objek.

b. Object

    Objek dibuat menggunakan konstruktor kelas,yang kemudian akan disebut turunan dari kelas.Instance dari kelas. Jika kelas secara umum merepresentasikan (template) sebuah object, maka sebuah instance adalah representasi nyata dari kelas itu sendiri
    
    Intinya adalah kelas itu lebih umum dan object lebih ke jenis-jenis dari kelas tersebut
    
   Contoh gambar definisi dari kelas dan object



    Keyword class di python biasanya lansung keyword class diikuti nama kelas yang diinginkan .
     - Init : Metode khusus yang digunakan untuk menginisialisasi pembuatan object dari kelas yang dibuat
     - Method : Setiap method harus memiliki parameter self yang artinya method tersebut dimiliki dan terdaftar                                 ke class tersebut untuk membedakan dari method atau fungsi yang ada di luar class
     - Mengakses Method biasanya menggunakan titik

 
   contoh kodingan Class vs Object

```python
class kubus :
    def __init__(self,s):
        self.sisi=s
    def tampilkan (self):
        print ("Kubus dengan sisi : ",self.sisi)
    def luas(self):
        l=(self.sisi*self.sisi)
        print("Luasnya : ",l)
    def keliling (self):
        k=4*(self.sisi)
        print("Kelilingnya  : ",k)
obj=kubus(8)
obj.tampilkan()
obj.luas()
obj.keliling()
```
